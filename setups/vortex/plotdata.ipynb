```python
path = '../../outputs/vortex/'
```

```python
ny = 128
nx = 128

x    = loadtxt(path+"domain_y.dat")[3:-3]
xmed = 0.5*(x[1:]+x[:-1])
```

```python
n = 500
sigma0 = fromfile(path+"gasdens0.dat" ).reshape(ny,nx)
sigma  = fromfile(path+"gasdens{:d}.dat".format(n) ).reshape(ny,nx)

vx0 = fromfile(path+"gasvy0.dat" ).reshape(ny,nx)
vx  = fromfile(path+"gasvy{:d}.dat".format(n) ).reshape(ny,nx)

vy0 = fromfile(path+"gasvx0.dat" ).reshape(ny,nx)
vy  = fromfile(path+"gasvx{:d}.dat".format(n) ).reshape(ny,nx)
```

```python
plot(xmed, sigma0[:,nx//2], 'o', mfc='none' )
plot(xmed, sigma[:,nx//2], '.', mfc='none' )

yscale('log')
show()
```

```python
plot(xmed, vy0[:,nx//2]+1.5*xmed, 'o', mfc='none' )
plot(xmed, vy[:,nx//2]+1.5*xmed, '.', mfc='none' )

show()
```

```python
plot(xmed, vy0[:,nx//2]- vy[:,nx//2])

show()
```
